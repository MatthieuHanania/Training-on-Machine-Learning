```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

df = pd.read_csv("length_of_stay.csv")

df = pd.get_dummies(df,columns=["condition","sex"])

df["log_bmi"] = np.log(1+df["bmi"])
df["weekly_sport_frequency_square"] = df["weekly_sport_frequency"]**2


y=df["length_of_stay"]
df = df.drop("length_of_stay",1)
X = df

X_train, X_test, y_train, y_test = train_test_split(X,y)

model1 = LinearRegression()

model1.fit(X_train,y_train)

mse_train = mean_squared_error(model1.predict(X_train),y_train)
mse_test = mean_squared_error(model1.predict(X_test),y_test)

print(mse_train, mse_test)

#########################################################################

model2 = BayesianRidge()

model2.fit(X_train,y_train)

mse_train = mean_squared_error(model2.predict(X_train),y_train)
mse_test = mean_squared_error(model2.predict(X_test),y_test)

print(mse_train, mse_test)

#########################################################################

model3 = DecisionTreeRegressor()

model3.fit(X_train,y_train)

mse_train = mean_squared_error(model3.predict(X_train),y_train)
mse_test = mean_squared_error(model3.predict(X_test),y_test)

print(mse_train, mse_test)

#########################################################################

model4 = RandomForestRegressor()

model4.fit(X_train,y_train)

mse_train = mean_squared_error(model4.predict(X_train),y_train)
mse_test = mean_squared_error(model4.predict(X_test),y_test)

print(mse_train, mse_test)

#########################################################################

model5 = SVR()

model5.fit(X_train,y_train)

mse_train = mean_squared_error(model5.predict(X_train),y_train)
mse_test = mean_squared_error(model5.predict(X_test),y_test)

print(mse_train, mse_test)

#########################################################################


```
